```python
import pandas as pd
```

```python
pd = pd.read_csv('analise_sentimento_final.csv')
```

```python
EMOTIONS = [
    'Não selecionado', 'Felicidade', 'Tristeza', 'Nojo',
    'Raiva', 'Medo', 'Surpresa', 'Desprezo', 'Neutro'
]
POLARITIES = ['Não selecionado', 'Positivo', 'Neutro', 'Negativo']
```

```python
def aplicando_filtro_ekman(sentimento):
    if sentimento == 'alegria':
        return 1
    elif sentimento == 'tristeza':
        return 2
    elif sentimento == 'nojo':
        return 3
    elif sentimento == 'raiva':
        return 4
    elif sentimento == 'medo':
        return 5
    elif sentimento == 'surpresa':
        return 6
    elif sentimento == 'desprezo':
        return 7
    else:
        return 8
```

```python
def aplicando_filtro_polaridade(polaridade):
    if polaridade == 'positivo':
        return 1
    elif polaridade == 'neutro':
        return 2
    elif polaridade == 'negativo':
        return 3
```

```python

# Aplicando os filtros no cabeçalho
pd['headline_sentiment'] = pd['headline_sentiment'].apply(aplicando_filtro_ekman)
pd['headline_polarity'] = pd['headline_polarity'].apply(aplicando_filtro_polaridade)

# Aplicando os filtros no texto
pd['sentence1_sentiment'] = pd['sentence1_sentiment'].apply(aplicando_filtro_ekman)
pd['sentence1_polarity'] = pd['sentence1_polarity'].apply(aplicando_filtro_polaridade)

pd['sentence2_sentiment'] = pd['sentence2_sentiment'].apply(aplicando_filtro_ekman)
pd['sentence2_polarity'] = pd['sentence2_polarity'].apply(aplicando_filtro_polaridade)

pd['sentence3_sentiment'] = pd['sentence3_sentiment'].apply(aplicando_filtro_ekman)
pd['sentence3_polarity'] = pd['sentence3_polarity'].apply(aplicando_filtro_polaridade)

# Aplicando os filtros no geral
pd['general_sentiment'] = pd['general_sentiment'].apply(aplicando_filtro_ekman)
pd['general_polarity'] = pd['general_polarity'].apply(aplicando_filtro_polaridade)

pd.head(5)
```
